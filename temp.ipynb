```python
def splitList(data):
    temp = data.str.strip().str.lower()

    ls = []

    for i in temp:
        try:
            [ls.append(k) for k in i.split(',')]
        except:
            pass

    ls = [x.strip(' ') for x in ls]

    return pd.Series(ls)


plotVal = splitList(CommonData['locallyAvailableFoodsConsumed'])
```

```python
import plotly.express as px
```

```python
fig = px.histogram(plotVal, title = 'Distribution of locallyAvailableFoodsConsumed', width = 1000)
```
